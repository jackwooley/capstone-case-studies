```python
import numpy as np
import pandas as pd
import random
```

```python
crime = pd.read_csv('crimedata.csv')
for i in range(0, crime.shape[0]):
    for j in range(0, crime.shape[1]):
        if crime.iloc[i, j] == '?':  # replace all '?' with an np.nan value (float type)
            crime.iloc[i, j] = np.nan
        elif 48 <= ord(str(crime.iloc[i, j])[0]) <= 57:  # if any string begins with the characters 0-9, convert it to a float type
            crime.iloc[i, j] = float(crime.iloc[i, j])
crime
```
